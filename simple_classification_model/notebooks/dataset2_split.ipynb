```python
import os
import random
import shutil
```

```python
healthy_dir = "/content/drive/MyDrive/bean_disease_project/dataset/healthy_1"
als_dir = "/content/drive/MyDrive/bean_disease_project/dataset/als_1"
bean_rust_dir = "/content/drive/MyDrive/bean_disease_project/dataset/bean_rust_1"
```

```python
def sample_data_files(src, dst, no_samples):
    for file in random.sample(os.listdir(src), no_samples):
        #check if file is jpg image
        if ".jpg" in file:
            file_path = os.path.join(src, file)
            #copy file
            print(f"Copying file {file_path} to {dst}")
            shutil.move(file_path, dst)
```

```python
train_healthy = "/content/drive/MyDrive/bean_disease_project/simple_classification_model/dataset2/train/healthy"
train_als = "/content/drive/MyDrive/bean_disease_project/simple_classification_model/dataset2/train/als"
train_bean_rust = "/content/drive/MyDrive/bean_disease_project/simple_classification_model/dataset2/train/bean_rust"

test_healthy = "/content/drive/MyDrive/bean_disease_project/simple_classification_model/dataset2/test/healthy"
test_als = "/content/drive/MyDrive/bean_disease_project/simple_classification_model/dataset2/test/als"
test_bean_rust = "/content/drive/MyDrive/bean_disease_project/simple_classification_model/dataset2/test/bean_rust"

val_healthy = "/content/drive/MyDrive/bean_disease_project/simple_classification_model/dataset2/validate/healthy"
val_als = "/content/drive/MyDrive/bean_disease_project/simple_classification_model/dataset2/validate/als"
val_bean_rust = "/content/drive/MyDrive/bean_disease_project/simple_classification_model/dataset2/validate/bean_rust"
```

```python
#healthy
sample_data_files(healthy_dir, train_healthy, 320)

#angular leaf spot
sample_data_files(als_dir, train_als, 320)

#bean rust
sample_data_files(bean_rust_dir, train_bean_rust, 320)
```

```python
#healthy
sample_data_files(healthy_dir, test_healthy, 40)

#angular leaf spot
sample_data_files(als_dir, test_als, 40)

#bean rust
sample_data_files(bean_rust_dir, test_bean_rust, 40)
```

```python
#healthy
sample_data_files(healthy_dir, train_healthy, 40)

#angular leaf spot
sample_data_files(als_dir, val_als, 40)

#bean rust
sample_data_files(bean_rust_dir, val_bean_rust, 40)
```
